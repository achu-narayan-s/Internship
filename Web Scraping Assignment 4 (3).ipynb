```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

try:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find('table', {'class': 'wikitable'})
    rows = table.find_all('tr')[1:]  # Exclude header row

    for row in rows:
        cells = row.find_all('td')
        rank = cells[0].text.strip()
        name = cells[1].text.strip()
        artist = cells[2].text.strip()
        upload_date = cells[4].text.strip()
        views = cells[5].text.strip()

        print("Rank:", rank)
        print("Name:", name)
        print("Artist:", artist)
        print("Upload Date:", upload_date)
        print("Views:", views)
        print('-' * 50)

except Exception as e:
    print("An error occurred:", e)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# Specify the URL
url = "https://www.bcci.tv/"

# Initialize a WebDriver
driver = webdriver.Chrome()

try:
    # Open the URL
    driver.get(url)
    
    # Dismiss the cookie popup if it appears
    try:
        cookie_popup = driver.find_element(By.CSS_SELECTOR, ".cookie.no-disply")
        if cookie_popup.is_displayed():
            cookie_popup.find_element(By.CSS_SELECTOR, ".js-accept-cookie").click()
    except:
        pass
    
    # Find the "International" link and scroll to it
    international_link = driver.find_element(By.XPATH, '//*[@id="navigation"]/ul[1]/li[2]/a')
    driver.execute_script("arguments[0].scrollIntoView();", international_link)
    time.sleep(1)  # Give some time for the page to scroll
    
    # Click the "International" link
    international_link.click()
    
    # Wait for the international fixtures to load
    time.sleep(5)
    # Adjust the waiting time as needed
    match_title=[]
    series=[]
    place=[]
    date=[]
    time=[]
    
    fixtures = driver.find_elements(By.CSS_SELECTOR, ".js-list-view .fixture__item")
    
    # Iterate through the fixtures and extract the details
    for fixture in fixtures:
        try:
            match_title = fixture.find_element(By.CLASS_NAME, "match-tournament-name ng-binding").text
            series = fixture.find_element(By.CSS_SELECTOR, ".fixture__description").text
            place = fixture.find_element(By.CSS_SELECTOR, ".fixture__location").text
            date = fixture.find_element(By.CSS_SELECTOR, ".fixture__date").text
            time = fixture.find_element(By.CSS_SELECTOR, ".fixture__time").text
            
            print("Match Title:", match_title)
            print("Series:", series)
            print("Place:", place)
            print("Date:", date)
            print("Time:", time)
            print('-' * 50)
        except Exception as e:
            print("An error occurred while extracting fixture details:", e)
    
except Exception as e:
    print("An error occurred:", e)





```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# Specify the URL
url = "http://statisticstimes.com/economy/gdp-of-indian-states.php"

# Initialize a WebDriver
driver = webdriver.Chrome()

try:
    # Open the URL
    driver.get(url)

    # Find the data table
    data_table = driver.find_element(By.XPATH, '//*[@id="table_id"]')

    # Get the HTML content of the data table
    table_html = data_table.get_attribute('outerHTML')
    print(table_html)  # Print the table HTML for inspection

    # Use BeautifulSoup to parse the HTML
    soup = BeautifulSoup(table_html, 'html.parser')

    # Find all rows in the table
    rows = soup.find_all('tr')
    print(rows)  # Print the rows for inspection

    # Extract and print the required details
    for row in rows[1:]:
        columns = row.find_all('td')
        rank = columns[0].text
        state = columns[1].text
        gsdp_18_19 = columns[2].text
        gsdp_19_20 = columns[3].text
        share_18_19 = columns[4].text
        gdp_billion = columns[5].text

        print("Rank:", rank)
        print("State:", state)
        print("GSDP(18-19) - at current prices:", gsdp_18_19)
        print("GSDP(19-20) - at current prices:", gsdp_19_20)
        print("Share(18-19):", share_18_19)
        print("GDP($ billion):", gdp_billion)
        print('-' * 50)

except Exception as e:
    print("An error occurred:", e)

finally:
    # Close the browser
    driver.quit()
```
