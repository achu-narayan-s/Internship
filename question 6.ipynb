```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.webdriver import WebDriver
import pandas as pd
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.get("https://www.flipkart.com/")
```

```python
search_keyword = driver.find_element(By.CLASS_NAME, "_3704LK")
search_keyword.clear()
search_keyword.send_keys('sneakers')
```

```python
search_keyword = driver.find_element(By.CLASS_NAME, "_34RNph")
search_keyword.click()
```

```python
brand_titles = []
product_types = []
product_prices = []
discount_provided = []
```

```python
title_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in title_tags[0:37]:
    title=i.text
    brand_titles.append(title)
```

```python
product_tags=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
for i in product_tags[0:37]:
    product=i.text
    product_types.append(product)
```

```python
prices_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
for i in prices_tags[0:37]:
    prices=i.text
    product_prices.append(prices)
```

```python
discounts_tags=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
for i in discounts_tags[0:37]:
    discounts=i.text
    discount_provided.append(discounts)
```

```python
print=(len(brand_titles),len(product_types),len(product_prices),len(discount_provided))
```

```python
print
```

```python
import pandas as pd
df=pd.DataFrame({'Title':brand_titles,'types':product_types,'prices':product_prices, 'discounts':discount_provided})
df
```
